#Giới thiệu chung về thư viện Matplotlib

Để thực hiện các suy luận thống kê cần thiết, cần phải trực quan hóa dữ liệu của bạn và Matplotlib là một trong những giải pháp như vậy cho người dùng Python. Nó là một thư viện vẽ đồ thị rất mạnh mẽ hữu ích cho những người làm việc với Python và NumPy.

###Tùy chỉnh về hình dạng và màu sắc

####Hình dạng đường

```python
#dot
ypoints = np.array([3, 10, 4, 9, 4, 5, 6, 9])
plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
#dash
ypoints = np.array([3, 10, 4, 9, 4, 5, 6, 9])
plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

```python
#dash and dot
ypoints = np.array([3, 10, 4, 9, 4, 5, 6, 9])
plt.plot(ypoints, linestyle = '-.')
plt.show()
```

####Kích thước đường

```python
ypoints = np.array([3, 10, 4, 9, 4, 5, 6, 9])
#Dùng thuộc tính linewidth để thay đổi kích thước của đường
plt.plot(ypoints, linewidth = '20')
plt.show()
```

####Màu sắc đường

```python
ypoints = np.array([3, 10, 4, 9, 4, 5, 6, 9])
#Thuộc tính color dùng để đổi màu đường thẳng
plt.plot(ypoints, color = 'hotpink')
plt.show()
```

```python
ypoints = np.array([3, 10, 4, 9, 4, 5, 6, 9])
#Thuộc tính color dùng để đổi màu đường thẳng
plt.plot(ypoints, c = '#4CAF50')
plt.show()
```

####Hình dạng điểm

```python
ypoints = np.array([3, 10, 4, 9, 4, 5, 6, 9])
#Muốn làm nổi bật điểm lên có thể sử dụng thêm marker
#marker o
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
ypoints = np.array([3, 10, 4, 9, 4, 5, 6, 9])
#Muốn làm nổi bật điểm lên có thể sử dụng thêm marker
#marker *
plt.plot(ypoints, marker = '*')
plt.show()
```

```python
ypoints = np.array([3, 10, 4, 9, 4, 5, 6, 9])
#Muốn làm nổi bật điểm lên có thể sử dụng thêm marker
#marker diamond
plt.plot(ypoints, marker = 'D')
plt.show()
```

####Kích thước điểm

```python
ypoints = np.array([3, 10, 4, 9, 4, 5, 6, 9])
#Tăng độ lớn của điểm bằng thuộc tính ms
plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

####Màu sắc điểm

```python
ypoints = np.array([3, 10, 4, 9, 4, 5, 6, 9])
#Dùng mec để đổi màu viền ngoài của marker
plt.plot(ypoints, marker = '*', ms = 20, mec = 'm')
plt.show()
```

```python
ypoints = np.array([3, 10, 4, 9, 4, 5, 6, 9])
#Dùng mfc để đổi màu của marker
plt.plot(ypoints, marker = 'H', ms = 20, mfc = 'r')
plt.show()
```

```python
#2 biển font1, font2 định nghĩa kiểu chữ, màu, kích thước. Là giá trị đầu vào của fontdict
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

##Hiển thị lưới (Grid)

##Biểu đồ dữ liệu phân tán (Scatter)

Biểu đồ phân tán được sử dụng để vẽ các điểm dữ liệu trên trục hoành và trục tung để thể hiện mức độ ảnh hưởng của một biến này bởi biến khác. Mỗi hàng trong bảng dữ liệu được biểu thị bằng một điểm đánh dấu, vị trí phụ thuộc vào giá trị của nó trong các cột được đặt trên trục X và Y. Một biến thứ ba có thể được đặt để tương ứng với màu sắc hoặc kích thước của các điểm đánh dấu, do đó thêm một chiều khác vào biểu đồ.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

```python
#Vẫn có thể thay đổi màu sắc, hình dạng và kích thước của mỗi điểm
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999')

plt.show()
```

```python
#Dùng 1 mảng để chứa màu sắc sau đó liên kết với 2 mảng
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

```python
#Dùng 1 color bar để thể hiện chiều sâu của dữ liệu, có thể áp dụng vào các dữ liệu như mật độ dân số để có cái nhìn trực quan về dữ liệu
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

##Biểu đồ tần suất (Histogram)

Biểu đồ tần suất hay biểu đồ phân phối tần suất là một dạng biểu đồ cột cho thấy bằng hình ảnh sự thay đổi, biến động của một tập hợp các dữ liệu theo những hình dạng nhất định

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

```python
import pandas as pd

# Generate data on commute times.
size, scale = 1000, 10
commutes = pd.Series(np.random.gamma(scale, size=size) ** 1.5)

commutes.plot.hist(grid=True, bins=20, rwidth=0.9,
                   color='#607c8e')
plt.title('Commute Times for 1,000 Commuters')
plt.xlabel('Counts')
plt.ylabel('Commute Time')
plt.grid(axis='y', alpha=0.75)
```

##Biểu đồ tròn (Pie)

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
#label cho các thành phần trong biểu đồ
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
#Chọn vị trí bắt đầu bằng startangle
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

```python
#Lấy ra 1 thành phần trong đồ thị
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0, 0.2, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0, 0, 0.2, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0.2]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
#Thêm bóng cho các thành phần 
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0.2]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
#Đối với nhiều biểu đồ phức tạp sử dụng label table để chú thích
y = np.array([30, 25, 10, 10, 5, 15, 5])
mylabels = ["Apples", "Bananas", "Cherries", "Dates", "Coconuts", "Pears", "Mangos"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```
